### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
# Calcular ingreso total por tienda
ingreso_tienda_1 = tienda["Precio"].sum()
ingreso_tienda_2 = tienda2["Precio"].sum()
ingreso_tienda_3 = tienda3["Precio"].sum()
ingreso_tienda_4 = tienda4["Precio"].sum()

# Mostrar resultados
print(f"Ingreso total Tienda 1: ${ingreso_tienda_1:,.2f}")
print(f"Ingreso total Tienda 2: ${ingreso_tienda_2:,.2f}")
print(f"Ingreso total Tienda 3: ${ingreso_tienda_3:,.2f}")
print(f"Ingreso total Tienda 4: ${ingreso_tienda_4:,.2f}")
```

# 2. Ventas por categoría

```python
# Limpiar nombres de columnas de todas las tiendas
tienda.columns = tienda.columns.str.strip()
tienda2.columns = tienda2.columns.str.strip()
tienda3.columns = tienda3.columns.str.strip()
tienda4.columns = tienda4.columns.str.strip()

ventas_categoria_1 = tienda["Categoría del Producto"].value_counts()
ventas_categoria_2 = tienda2["Categoría del Producto"].value_counts()
ventas_categoria_3 = tienda3["Categoría del Producto"].value_counts()
ventas_categoria_4 = tienda4["Categoría del Producto"].value_counts()

print("Tienda 1 - Ventas por categoría:")
print(ventas_categoria_1, end="\n\n")

print("Tienda 2 - Ventas por categoría:")
print(ventas_categoria_2, end="\n\n")

print("Tienda 3 - Ventas por categoría:")
print(ventas_categoria_3, end="\n\n")

print("Tienda 4 - Ventas por categoría:")
print(ventas_categoria_4, end="\n\n")
```

# 3. Calificación promedio de la tienda


```python
# Limpia espacios/tabs de los nombres de columnas
tienda.columns = tienda.columns.str.strip()
tienda2.columns = tienda2.columns.str.strip()
tienda3.columns = tienda3.columns.str.strip()
tienda4.columns = tienda4.columns.str.strip()

# Calcula promedio de calificaciones
promedio_calif_1 = tienda["Calificación"].mean()
promedio_calif_2 = tienda2["Calificación"].mean()
promedio_calif_3 = tienda3["Calificación"].mean()
promedio_calif_4 = tienda4["Calificación"].mean()

# Mostrar resultados
print(f"Calificación promedio Tienda 1: {promedio_calif_1:.2f}")
print(f"Calificación promedio Tienda 2: {promedio_calif_2:.2f}")
print(f"Calificación promedio Tienda 3: {promedio_calif_3:.2f}")
print(f"Calificación promedio Tienda 4: {promedio_calif_4:.2f}")
```

# 4. Productos más y menos vendidos

```python
# Limpiar nombres de columnas
tienda.columns = tienda.columns.str.strip()
tienda2.columns = tienda2.columns.str.strip()
tienda3.columns = tienda3.columns.str.strip()
tienda4.columns = tienda4.columns.str.strip()


# Productos más y menos vendidos
mas_vendidos_1 = tienda["Producto"].value_counts().head(5)
menos_vendidos_1 = tienda["Producto"].value_counts().tail(5)

mas_vendidos_2 = tienda2["Producto"].value_counts().head(5)
menos_vendidos_2 = tienda2["Producto"].value_counts().tail(5)

mas_vendidos_3 = tienda3["Producto"].value_counts().head(5)
menos_vendidos_3 = tienda3["Producto"].value_counts().tail(5)

mas_vendidos_4 = tienda4["Producto"].value_counts().head(5)
menos_vendidos_4 = tienda4["Producto"].value_counts().tail(5)


import matplotlib.pyplot as plt

def graficar_top_bottom(mas, menos, titulo):
    fig, axs = plt.subplots(1, 2, figsize=(14, 5))
    mas.plot(kind='bar', ax=axs[0], color='green')
    menos.plot(kind='bar', ax=axs[1], color='red')
    axs[0].set_title(f"Top productos más vendidos - {titulo}")
    axs[1].set_title(f"Productos menos vendidos - {titulo}")
    axs[0].set_ylabel("Cantidad Vendida")
    axs[1].set_ylabel("Cantidad Vendida")
    plt.tight_layout()
    plt.show()

# Graficar cada tienda
graficar_top_bottom(mas_vendidos_1, menos_vendidos_1, "Tienda 1")
graficar_top_bottom(mas_vendidos_2, menos_vendidos_2, "Tienda 2")
graficar_top_bottom(mas_vendidos_3, menos_vendidos_3, "Tienda 3")
graficar_top_bottom(mas_vendidos_4, menos_vendidos_4, "Tienda 4")
```
